```python
snacks['fred'] = snacks.number_of_snacks.where(snacks.snack_fred == True, 0)
snacks['daphnie'] = snacks.number_of_snacks.where(snacks.snack_daphnie == True, 0)
snacks['velma'] = snacks.number_of_snacks.where(snacks.snack_velma == True, 0)
snacks['shaggy'] = snacks.number_of_snacks.where(snacks.snack_shaggy == True, 0)
snacks['scooby'] = snacks.number_of_snacks.where(snacks.snack_scooby == True, 0)
```

```python
li_fred = []
li_daphn = []
li_velma = []
li_shaggy = []
li_scooby = []
for i in range(len(snacks)):
    li_fred.append(snacks.fred[i])
    li_daphn.append(snacks.daphnie[i])
    li_velma.append(snacks.velma[i])
    li_shaggy.append(snacks.shaggy[i])
    li_scooby.append(snacks.scooby[i])
```

```python
res = pd.DataFrame(columns=['Names','Values'])

new_row1 = {'Names':'Fred', 'Values':li_fred}
new_row2 = {'Names':'Daphnie', 'Values':li_daphn}
new_row3 = {'Names':'Velma', 'Values':li_velma}
new_row4 = {'Names':'Shaggy', 'Values':li_shaggy}
new_row5 = {'Names':'Scooby', 'Values':li_scooby}

res = res.append(new_row1, ignore_index=True)
res = res.append(new_row2, ignore_index=True)
res = res.append(new_row3, ignore_index=True)
res = res.append(new_row4, ignore_index=True)
res = res.append(new_row5, ignore_index=True)

res = res.explode('Values')
res['Values'] = res['Values'].astype('int')
```

```python
font = {'weight':'ultralight',
        'size'   : 20}

matplotlib.rc('font', **font)
plt.figure(figsize=(30, 20))

ax = sns.violinplot(x='Names', y="Values", data=res,  scale = 'width', width = 0.9)
plt.savefig('violins.png')
```
