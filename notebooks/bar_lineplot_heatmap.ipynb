```python
import pandas as pd
import matplotlib.pyplot as plt
data_url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-13/scoobydoo.csv'
data_df = pd.read_csv(data_url)
data_df
```

```python
unmask_fred = len(data_df[data_df["unmask_fred"] == True])
unmask_daphnie = len(data_df[data_df["unmask_daphnie"] == True])
unmask_velma = len(data_df[data_df["unmask_velma"] == True])
unmask_shaggy = len(data_df[data_df["unmask_shaggy"] == True])
unmask_scooby = len(data_df[data_df["unmask_scooby"] == True])
caught_fred = len(data_df[data_df["caught_fred"] == True])
caught_daphnie = len(data_df[data_df["caught_daphnie"] == True])
caught_velma = len(data_df[data_df["caught_velma"] == True])
caught_shaggy = len(data_df[data_df["caught_shaggy"] == True])
caught_scooby = len(data_df[data_df["caught_scooby"] == True])
```

```python
import seaborn as sns
df = pd.DataFrame({'unmask_fred':[unmask_fred], 'unmask_daphnie':[unmask_daphnie],
                   'unmask_velma':[unmask_velma], 'unmask_shaggy':[unmask_shaggy],
                   'unmask_scooby':[unmask_scooby],})

plt.figure(figsize=(8, 3))
plt.bar(['fred', 'daphnie', 'velma', 'shaggy', 'scooby'], [unmask_fred, unmask_daphnie, unmask_velma, unmask_shaggy, unmask_scooby], color='g')

def autolabel(rects, labels=None, height_factor=1.01):
    for i, rect in enumerate(rects):
        height = rect.get_height()
        if labels is not None:
            try:
                label = labels[i]
            except (TypeError, KeyError):
                label = ' '
        else:
            label = '%d' % int(height)
        ax.text(rect.get_x() + rect.get_width()/2., height_factor*height,
                '{}'.format(label),
                ha='center', va='bottom')

ax = plt.gca()
autolabel(ax.patches, height_factor=0.99)
plt.title('Unmasked by')

plt.savefig('unmasked by.png')
```

```python
import seaborn as sns
df = pd.DataFrame({'unmask_fred':[caught_fred], 'unmask_daphnie':[caught_daphnie],
                   'unmask_velma':[caught_velma], 'unmask_shaggy':[caught_shaggy],
                   'unmask_scooby':[caught_scooby],})

plt.figure(figsize=(8, 3))
plt.bar(['fred', 'daphnie', 'velma', 'shaggy', 'scooby'], [caught_fred, caught_daphnie, caught_velma, caught_shaggy, caught_scooby], color='y')
ax = plt.gca()
autolabel(ax.patches, height_factor=0.99)
plt.title('Caught by')
plt.savefig('Caught by.png')
```
