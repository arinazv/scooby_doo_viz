This notebook shows the process of making a box and whiskers plot. Firstly, we did a plot based on all of the series. But as you can see the result was not really readable. And that is why we decided to base our chart only on the first five series. 
The plot itself represents how the ratings of different series differ. 

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data_url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-13/scoobydoo.csv'
scooby = pd.read_csv(data_url)
```

```python
box = scooby[['series_name', 'season', 'imdb' ]]
```

```python
plt.figure(figsize=(20, 15))
ax = sns.boxplot(x='series_name', y="imdb", data=box, width = 1)
```

```python
box1 = box.loc[box['series_name'] == 'Scooby Doo, Where Are You!']
```

```python
box2 = box.loc[box['series_name'] == 'The New Scooby-Doo Movies']
```

```python
box3 = box.loc[box['series_name'] == 'The Scooby-Doo Show']
```

```python
box4 = box.loc[box['series_name'] == 'Laff-a-Lympics']
```

```python
box5 = box.loc[box['series_name'] == 'Scooby-Doo and Scrappy-Doo (second series)']
```

```python
box6 = box.loc[box['series_name'] == 'A Pup Named Scooby-Doo']
```

```python
res = pd.concat([box1, box2, box3, box4, box5, box6], axis=0)
```

```python
plt.figure(figsize=(20, 15))

font = {'weight':'ultralight',
        'size'   : 10}
matplotlib.rc('font', **font)

ax = sns.boxplot(x='series_name', y="imdb", data=res)
ax.set_xlabel('Name of the Series', fontsize=16)
ax.set_ylabel('IMDB Score', fontsize=16)

plt.savefig('boxplot.png')
```
